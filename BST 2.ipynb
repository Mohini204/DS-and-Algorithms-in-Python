```python
class BinaryTreeNode:
    
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
```

#### Levelwise Input

```python
import queue

def levelwiseInput():
    
    q = queue.Queue()
    print('Enter root ')
    rootdata = int(input())
    if rootdata == -1:
        return None
    
    root = BinaryTreeNode(rootdata)
        
    q.put(root)
    
    while not q.empty():
        
        curr = q.get()
        
        print('Enter left of ', curr.data)
        leftchilddata = int(input())
        if leftchilddata != -1:
            leftchild = BinaryTreeNode(leftchilddata)
            curr.left = leftchild
            q.put(curr.left)
            
        print('Enter right of ',curr.data)
        rightchilddata = int(input())
        if rightchilddata != -1:
            rightchild = BinaryTreeNode(rightchilddata)
            curr.right = rightchild
            q.put(curr.right)
            
    return root
        
```

```python
import queue

def levelwisePrint(root):
    
    q = queue.Queue()
    
    if root is None:
        return
    
    q.put(root)
    
    while not q.empty():
        
        curr = q.get()
        print(curr.data, end=':')
        
        if curr.left is not None:
            print('L:',end='')
            print(curr.left.data, end = ',')
            q.put(curr.left)
        else:
            print('L:',end='')
            print(-1, end = ',')
        if curr.right is not None:
            print('R:',end='')
            print(curr.right.data, end = '')
            q.put(curr.right)
        else:
            print('R:',end = '')
            print(-1,end='')
        print()  
```

#### Root to Node Path in Binary Tree

```python
def findPathBST(root,data):
    
    if root is None:
        return None
    
    if root.data == data:
        l = []
        l.append(root.data)
        return l
    
    la = findPathBST(root.left,data)
    ra = findPathBST(root.right,data)
    
    if la is not None:
        la.append(root.data)
        return la
    
    if ra is not None:
        ra.append(root.data)
        return ra
    
    return None
```

```python
class BST:
    
    def __init__(self):
        self.root = None
        self.numNodes = 0
        
    def printTreeHelper(self,root):
        if root == None:
            return
        
        print(root.data, end=':')
        if root.left != None:
            print('L',root.left.data, end=',')
        if root.right != None:
            print('R',root.right.data, end='')    
        print()
        
        self.printTreeHelper(root.left)
        self.printTreeHelper(root.right)
        
        
    def printTree(self):
        self.printTreeHelper(self.root)
    
    def isPresentHelper(self,root,data):
    
        if root is None:
            return False
        
        if root.data == data:
            return True
        
        if root.data > data:
            return self.isPresentHelper(root.left,data)
        
        if root.data < data:
            return self.isPresentHelper(root.right,data)
        
    def isPresent(self, data):
        return self.isPresentHelper(self.root,data)
    
    def count(self):
        return self.numNodes
    
    def insertHelper(self,root,data):
        
        if root is None:
            newNode = BinaryTreeNode(data)
            return newNode
        
        if root.data > data:
            root.left = self.insertHelper(root.left,data)
            return root
        else:
            root.right = self.insertHelper(root.right,data)
            return root
    
    def insert(self,data):
        self.numNodes+=1
        self.root = self.insertHelper(self.root,data)
        return
    
    def minTree(root):
    
        if root is None:
            return 10000
    
        if root.left is None:
            return root.data
        
        return self.minTree(root.left)
    
    def deleteDataHelper(root,data):
        
        if root is None:
            return False, None
        
        if root.data > data:
            deleted, newleft = self.deleteDataHelper(root.left, data)
            root.left = newleft
            return deleted, root
        
        elif root.data < data:
            deleted, newright = self.deleteDataHelper(root.right, data)
            root.right = newright
            return deleted, root
        
        else:
            #root is leaf
            if root.left is None and root.right is None:
                return True, None
            
            #root has only right child
            elif root.left is None and root.right is not None:
                return True, root.right
            
            #root has only left child
            elif root.left is not None and root.right is None:
                return True, root.left
            
            #root has both children
            else:
                minRight = self.minTree(root)
                root.data = minRight
                deleted, newright = self.deleteDataHelper(root.right, minRight)
                root.right = newright
                return True, root
    
    def deleteData(self,data):
        deleted, newroot = self.deleteDataHelper(self.root,data)
        if deleted:
            self.numNodes-=1
        self.root = newroot
        return deleted
```
